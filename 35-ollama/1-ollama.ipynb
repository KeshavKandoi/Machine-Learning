```python
import base64
import ollama

image_path="KK.png"

with open(image_path,"rb") as f:
  image_bytes=f.read()
image_64=base64.b64encode(image_bytes).decode("utf-8") 

response= ollama.generate(model="qwen3.5", images=[image_64], prompt="Give caption for the image.")
print(response.response)
```

```python
# PASSING MULTIPLE IMAGES TO THE MODEL.

image_paths = ["KK.png","KK2.png"]

images_base64=[]
for i in image_paths:
    with open(i , "rb") as f:
        image_bytes= f.read()
        images_base64.append(base64.b64encode(image_bytes).decode("utf-8"))

response= ollama.generate(model="qwen3.5", images=images_base64,
                          prompt="Generate an story based on these images, make sure you take context from each and every image.")
print(response.response)
```

```python
# SYSTEM INSTRUCTIONS
response = ollama.generate(model="qwen3.5",prompt="why does moon glow ?", system="You are an funny assistant , you explain things in funny way")
print(response.response)
```
