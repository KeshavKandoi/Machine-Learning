```python
df_customers=pd.DataFrame(
  {
    "customer_id":[1,2,3,4],
    "name":["Adam","Bob","Charlie","Dave"]
  }
)
df_customers
```

```python
df_orders=pd.DataFrame(
  {
    "order_id":[101,102,103,104],
    "customer_id":[2,1,4,5],
    "amount":[250,120,300,180],
   
  }
)
df_orders
```

```python
pd.merge(df_customers,df_orders,on="customer_id")#innerjoin
```

```python
pd.merge(df_customers,df_orders,on="customer_id",how="left")#leftjoin
```

```python
pd.merge(df_customers,df_orders,on="customer_id",how="right")#rightjoin
```

```python
pd.merge(df_customers,df_orders,on="customer_id",how="outer")#outerjoin
```
