```python
import json
```

```python
def load_data(filename):
  with open(filename,"r") as f:
    data=json.load(f)
  return data
```

```python
data=load_data("store_data.json")
print(data)
```

```python
# Clean & structure the data
def clean_data(data):
    text_to_num = {
        "one": 1,
        "two": 2,
        "three": 3,
        "four": 4,
        "five": 5
    }

    cleaned_data = []
    unique_users = set()

    for user in data:

        # -----------------------------
        # Clean ratings (data consistency)
        # -----------------------------
        raw_rating = str(user.get("rating")).strip().lower()

        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]

        user["rating"] = raw_rating

        # -----------------------------
        # Handle missing values
        # -----------------------------
        raw_age = user.get("age")
        if raw_age is None:
            user["age"] = None

        # -----------------------------
        # Deduplication (by name)
        # -----------------------------
        name = user.get("name", "").strip().lower()
        if name in unique_users:
            continue

        unique_users.add(name)
        cleaned_data.append(user)

    return cleaned_data
```

```python
clean_data(data)
```

```python
def get_insight(data):

  tot_rating=0
  for user in data:
    tot_rating+=float(user["rating"])
  print(f"avg rating={tot_rating/len(data)}")

  poor_rating=0
  for user in data:
    if(float(user["rating"])<5):
      poor_rating+=1
  print(f"% of user with avg rating={poor_rating/len(data)*100}")

```

```python
get_insight(data)
```

```python
def get_recommendation(data):
  recommendation = []
  for user in data:
    curr_recom={}
    curr_recom["name"]=user["name"]

    if(float(user["rating"])>=4):
      curr_recom["brand"]="Apple"
    else:
      curr_recom["brand"]="samsung"
  
    recommendation.append(curr_recom)

  return recommendation
   
     



  
```
